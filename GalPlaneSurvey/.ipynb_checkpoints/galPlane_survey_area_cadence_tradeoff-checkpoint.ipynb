## Exploration of the trade-offs between Galactic Science survey area and cadence

Given the limitations of available on-sky time, there is an inevitable trade-off between the sky area that can be surveyed and the cadence which can be achieved over that area.  

The purpose of this notebook is to explore this trade off and identify a "sweet spot" if possible.

### Assumptions:
* The cadence within the galactic science minisurvey area has to be the same for all HEALpixels
* The same survey footprint must be observed in all 10yrs of the survey
* The balance of time between the different filters is constant, meaning that equal numbers of exposures are obtained at the same cadence in all 6 LSST filters. 
* The total time allocated to the galactic science minisurvey is fixed.  
    * The total time allocated is handled as a fixed total number of visits by Rubin Observatory to HEALpixels within the galactic science region of interest.  This avoids the complexities of slew time etc, which are handled by the scheduler, and will be optimized separately. 
    * The total number of visits, NTotalVisits, is TBD, and the starting guess value used here is based on previous LSST simulations. 
* The sky region of interest to galactic science is drawn from previous work to build the HEALpixel priority sky maps for each LSST filter.  
    * The "combined" region refers to the region of the sky where the assigned priority is > 0.0 once all galactic science use-cases are combined.

```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d
%matplotlib inline
import rubin_sim.maf as maf
from rubin_sim.data import get_data_dir
import healpy as hp
from astropy import units as u
from astropy_healpix import HEALPix
from astropy.coordinates import Galactic, TETE, SkyCoord
from astropy.io import fits

NSIDE = 64
NPIX = hp.nside2npix(NSIDE)

# Total number of visits to the combined galactic science region of interest:
NTotalVisits = 2088400

# Time taken per visit [in seconds] to a single field (estimated, taken from https://lsst.org/scientists/keynumbers):
TimePerVisit = 42.2

# Area surveyed in a single visit [sq deg]:
AreaPerVisit = 9.6
```

If we assume a fixed and constant survey cadence to each HEALpixel within the survey footprint, then the maximum total area that can be surveyed at that cadence can be calculated by the following function.  

```python
def calcSurveyArea(NTotalVisits, AreaPerVisit, cadence):
    """
    Function to calculate the total area which can be surveyed at the given cadence 
    [in days, assumed to be greater than 1day], given the total number of available visits.  
    As LSST will continue for 10yrs, this has to take into account the average seasonal 
    visibility.
    
    NTotalVisits - int, dimensionless - Total number of visits to the whole survey region over 10yrs
    TimePerVisit - float, units: seconds - Time taken for a single visit to a single field pointing
    cadence - array of floats, units: days - Expected interval between sequential visits to any given pointing 
    """
    
    if cadence.any() < 1.0:
        raise IOError('This function assumes the minimum cadence achieved will be greater than once per night')
        
    # Assuming that any given field can be observed for 30d per month for 
    # 6 months of the year.  
    nightsObsSeason = 6.0 * 30.0
    nVisitsPerFieldPerYear = nightsObsSeason / cadence
    
    # Since we have to revisit the same survey area every year over the 10yr survey, 
    # calculate the total visits available per year
    nVisitsPerYear = NTotalVisits / 10.0
    
    # This figure is further reduced by the scheduler's dithering function, which effectively means
    # more visits are required per pointing to ensure that that HEALpixel is fully observed.
    # Rather than re-implement the scheulder, we simulate that here by reducing the available
    # number of visits.
    # WARNING: this factor is approximate
    nVisitsPerYear = nVisitsPerYear / 6
    
    # Calculate the total area that can be surveyed each year at this cadence
    nPointingsPerYear = nVisitsPerYear / nVisitsPerFieldPerYear
    areaSurveyed = nPointingsPerYear * AreaPerVisit
    
    return areaSurveyed
```

```python
cadences = np.arange(1.0,20.0,1.0)
areas = calcSurveyArea(NTotalVisits, AreaPerVisit, cadences)

(fig, ax) = plt.subplots(1, 1, figsize=(10, 10))
plt.plot(cadences, areas, 'r-')
plt.xlabel('Cadence interval [days]', fontsize=16)
plt.ylabel('Area surveyed per year [deg$^{2}$]', fontsize=16)
plt.xticks(fontsize=16)
plt.yticks(fontsize=16)
```

The galactic science sky maps assign a priority to each HEALpixel in the sky according to the combined range of science goals.  If we assume that the HEALpixels to be observed first are the highest priority ones, we can use these priority maps together with the linear relationship above to calculate the total scientific priority of the area surveyed per year as a function of cadence.  

This is directly analogous to the GalacticPlaneFootprintMetric. 

```python
# Load the galactic science priority map data from the "sum" file
home = os.path.expanduser("~")
galSciencePriorityMapFile = os.path.join(home, 'rubin_sim_data/maps/GalacticPlanePriorityMaps/priority_GalPlane_footprint_map_data_sum.fits')

with fits.open(galSciencePriorityMapFile) as hdul:
    galSciencePriorityMapTable = hdul[1].data

# The sum priority map sums the priority assigned to all HEALpixels over all 6 LSST filters.  
# For ease of understanding, it is valid to average this over all filters here.  
# We also select the combined region map, as opposed to the maps for the distinct science cases
priorityMap = getattr(galSciencePriorityMapTable, 'combined_map')
priorityMap /= 6

# Select submap data, if desired:
priorityThreshold = 0.5
idx = np.where(priorityMap > priorityThreshold)[0]
print('Number of HEALpixels in desired survey region with priority > '+str(priorityThreshold)+': '+str(len(idx)))
selectedMap = np.zeros(len(priorityMap))
selectedMap[idx] = priorityMap[idx]

# For illustration, plot the full map data:
hp.mollview(selectedMap, title='Galactic science region of interest')
```

```python
def prioritizeSurveyPixels(priorityMap):
    """
    Function to sort the HEALpixels in the desired survey area into order of decending priority.
    """

    pixOrder = np.argsort(priorityMap)
    pixOrder = pixOrder[::-1]
    mapPixPriority = priorityMap[pixOrder]

    return mapPixPriority
```

```python
def PriorityOfSurveyArea(NSIDE, surveyArea, mapPixPriority):
    """
    Function calculates the sum of the priorities assigned to all HEALpixels within a given survey area, 
    assuming that the highest priority pixels would be selected first. 
    Returns:
        areaPriority - the sum of the priority of all included pixels
        minPriority - the minimum value of HEALpixel priority included in the region
    """
    
    pixArea = hp.nside2pixarea(NSIDE, degrees = True)
    nPixels = (surveyArea / pixArea).astype(int)
    areaPriority = np.zeros(len(surveyArea))
    minPriority = np.zeros(len(surveyArea))
    for i in range(0,len(surveyArea),1):
        areaPriority[i] = mapPixPriority[:nPixels[i]].sum()
        minPriority[i] = min(mapPixPriority[:nPixels[i]])
        
    return areaPriority, minPriority
```

First we pre-sort the HEALpixels within the galactic science survey region into priority order, to avoid doing this repeatedly later on

```python
mapPixPriority = prioritizeSurveyPixels(priorityMap)
```

So now we can modified the plot above to chart the metric value as a function of the area / cadence interval trade-off

```python
cadences = np.arange(1.0,20.0,1.0)
areas = calcSurveyArea(NTotalVisits, AreaPerVisit, cadences)
(areaPriority,minPriority) = PriorityOfSurveyArea(NSIDE, areas, mapPixPriority)

(fig, ax1) = plt.subplots(1, 1, figsize=(10, 10))
ax1.plot(cadences, areaPriority, 'r-')
ax2 = ax1.twiny()
ax2.plot(areas, areaPriority, 'r-')
ax3 = ax1.twinx()
ax3.plot(cadences, minPriority, 'b-.')

ax1.set_xlabel('Cadence interval [days]', fontsize=16)
ax1.set_ylabel('Total priority of the surveyed area', fontsize=16, color='r')
ax1.tick_params(axis='x', labelsize=16)
ax1.tick_params(axis='y', labelsize=16)
ax2.tick_params(axis='x', labelsize=16)
ax2.set_xlabel('Area surveyed per year [deg$^{2}$]', fontsize=16)
ax3.tick_params(axis='y', labelsize=16)
ax3.set_ylabel('Priority selection threshold', fontsize=16, color='b')
```

A few conclusions may be drawn from this graph:
* Virtually the entire galactic science region of interest can be surveyed with a uniform ~10d cadence interval 
* The plateau in the priority selection threshold ~0.14 results from the "dusty plane" region, which is relatively low-priority for galactic science as well as the WFD, but covers a large area, so including it produces a marked penalty on the cadence achievable for the rest of the survey region
* A priority selection threshold of 0.4 means that a small region of the very highest priority areas can be surveyed with a near-1d cadence suitable for science at all timescales

CAVEAT: This depends strongly on the dithering function applied, meaning that the cadence realised in practice will probably be lower.

## High and low cadence regions in the Galactic Plane

A uniform 1d cadence across the entire galactic science survey area may not necessarily be the best use of time.  
Based on the science use-cases included, there are limited regions for which such a high cadence is necessary, including:
* The Galactic Bulge, in particular the Roman Bulge survey field for planetary/stellar microlensing science
* The Magellanic Clouds, for microlensing science
* The Bonito Star Forming Regions, for Young Stellar Object science
* Open and Globular clusters
* Selected pencilbeams, for pulsating stars, sampling a range of galactic structure

If the scheduler software can be adapted to cover these limited regions at high (about 2d) cadence, the rest of the galactic science survey region could be covered at lower (about 5d) cadence and still achieve the majority of the desired science. 


### Designing a high-cadence region


The combined galactic science priority map currently includes two very wide area components, representing the distribution of stars across the Galactic Plane and the distribution of X-ray binaries, which is related to the former. These can be considered to indicate the default priority of a given HEALpixel.  

On top of that, the map includes a number of regions of special scientific interest, plus the pencilbeam fields.  

To begin to design a high-cadence region, we start with the just the regions of special scientific interest, and explore total survey area this represents.

```python
def combined_pixel_regions(galSciencePriorityMapTable, regionList, priorityThreshold=0.0):
    combinedPixels = set([])
    for region in regionList:
        pixels = np.where(galSciencePriorityMapTable[region]/6.0 > priorityThreshold)[0]
        combinedPixels = combinedPixels.union(set(pixels))
    combinedPixels = list(combinedPixels)

    return combinedPixels
```

```python
highCadenceRegions = ['galactic_bulge_map','magellenic_clouds_map','bonito_sfr_map','clementini_stellarpops_map','open_clusters_map','globular_clusters_map']
#highCadenceRegions = ['clementini_stellarpops_map']

highCadencePixels = combined_pixel_regions(galSciencePriorityMapTable, highCadenceRegions)

highCadenceRegion = np.zeros(NPIX)
highCadenceRegion[highCadencePixels] = galSciencePriorityMapTable['combined_map'][highCadencePixels]

hp.mollview(highCadenceRegion, title='Galactic science region - all sky')
```

This has to be combined with a map of the Rubin visibility zone, since the maps are built from all-sky catalogs and include regions that cannot be observed from the Rubin site.

```python
home = os.path.expanduser("~")
visibilityMapFile = os.path.join(home, 'software','LSST-TVS_software_tools','footprint_maps','baseline_v2.0_10yrs_rubin_visibility_zone.fits')
print(visibilityMapFile)

with fits.open(visibilityMapFile) as hdul:
    visibilityMap = hdul[1].data['visibility_map']
```

```python
visiblePixels = np.where(visibilityMap > 0.0)[0]
highCadencePixels = list(set(highCadencePixels).intersection(set(visiblePixels)))

highCadenceRegion = np.zeros(NPIX)
highCadenceRegion[highCadencePixels] = galSciencePriorityMapTable['combined_map'][highCadencePixels]

hp.mollview(highCadenceRegion, title='Galactic science high-cadence region')
```

```python
pixArea = hp.nside2pixarea(NSIDE, degrees = True)
highCadenceArea = len(highCadencePixels)*pixArea
print('Number of high-cadence HEALpixels: '+str(len(highCadencePixels)))
print('Total area of high cadence HEALpixels: '+str(highCadenceArea)+' sq.deg.')
nRubinPointings = round(highCadenceArea / AreaPerVisit, 0)
print('Total number of Rubin fields: '+str(nRubinPointings))
```

### IMPORTANT

This proposed high-priority region is preliminary, and it is important to note that there is scope for considerable optimization of this high cadence zone region.  

Instead, we propose to use the above map to optimize the locations of the pencilbeams towards fewer but larger, contiguous regions that incorporate key clusters, SFRs of interest.  
In the process, we will also explore:
* Overlap with the K2 fields
* Selection of priority cluster fields (this map simply incorporates all of them)

This optimization is a work in progress by the community, and will substantially revise the above map to condense it into fewer, distinct regions.  

We continue the current exploration with the above map for the time being for the sake of reference. 

The total size of this region suggests that it may be feasible to survey it at high cadence, while leaving a significant number of visits, which could be dedicated to monitoring the other regions of interest to galactic science at lower cadence. 

### Designing a low-cadence region

We can follow a similar procedure to identify a low-cadence region, this time based on the Galactic Plane and X-ray binary distribution maps that are both functions of stellar density.  

```python
def combined_pixel_distributions(galSciencePriorityMapTable, regionList, priorityThresholds):
    combinedPixels = set([])
    for i,region in enumerate(regionList):
        pixels = np.where(galSciencePriorityMapTable[region]/6 > priorityThresholds[i])[0]
        combinedPixels = combinedPixels.union(set(pixels))
    combinedPixels = list(combinedPixels)

    return combinedPixels
```

```python
def defineLowCadenceRegion(galSciencePriorityMapTable, visiblePixels, highCadencePixels, priorityThresholds=[0.2, 0.00005]):
    lowCadenceRegions = ['galactic_plane_map','xrb_priority_map']
    lowCadencePixels = combined_pixel_distributions(galSciencePriorityMapTable, lowCadenceRegions, priorityThresholds)

    # Factor in Rubin visibility...
    lowCadencePixels = list(set(lowCadencePixels).intersection(set(visiblePixels)))

    # But to avoid duplicating coverage of the high-cadence regions, we should exclude those pixels...
    lowCadencePixels = list(set(lowCadencePixels).difference(set(highCadencePixels)))

    lowCadenceRegion = np.zeros(NPIX)
    lowCadenceRegion[lowCadencePixels] = galSciencePriorityMapTable['galactic_plane_map'][lowCadencePixels] + \
                                            galSciencePriorityMapTable['xrb_priority_map'][lowCadencePixels]

    return lowCadenceRegion

hp.mollview(lowCadenceRegion, title='Galactic science low-cadence region')
```

```python
pixArea = hp.nside2pixarea(NSIDE, degrees = True)
lowCadenceArea = len(lowCadencePixels)*pixArea
print('Number of low-cadence HEALpixels: '+str(len(lowCadencePixels)))
print('Total area of low cadence HEALpixels: '+str(lowCadenceArea)+' sq.deg.')
nRubinPointings = round(lowCadenceArea / AreaPerVisit, 0)
print('Total number of Rubin fields: '+str(nRubinPointings))
```

## Allocating visits to the high- and low- survey regions

There is a trade-off to explore between the cadence allocated to the high and low cadence regions, and the priority thresholds used to select the region areas, which we explore here. 

```python
def estimate_survey_low_cadence(highCadencePixels,lowCadencePixels,highCadence,NTotalVisits,verbose=True):
    """
    Function to estimate the achievable cadence in the low cadence survey region, given a known cadence for the 
    high-cadence region
    """
    
    if verbose: print('High cadence requested: ',highCadence)
    
    # Since we have to revisit the same survey area every year over the 10yr survey, 
    # calculate the total visits available per year
    nVisitsPerYear = NTotalVisits / 10.0
    
    # This figure is further reduced by the scheduler's dithering function, which effectively means
    # more visits are required per pointing to ensure that that HEALpixel is fully observed.
    # Rather than re-implement the scheulder, we simulate that here by reducing the available
    # number of visits.
    # WARNING: this factor is approximate
    ditherFactor = 6
    
    # Number of visits required to survey the high cadence region at the cadence requested
    pixArea = hp.nside2pixarea(NSIDE, degrees = True)
    highCadenceArea = len(highCadencePixels)*pixArea
    nHighPointings = round(highCadenceArea / AreaPerVisit, 0)
    if verbose: print('Number of pointings in high cadence region: ',nHighPointings)
    
    # Figure out how many times we need to revisit each high cadence pointing in an observing season
    nightsObsSeason = 6.0 * 30.0
    nVisitsPerFieldPerYear = nightsObsSeason / highCadence
    if verbose: print('Number of visits per pointing per year at this cadence: ',nVisitsPerFieldPerYear)
    
    # The total number of visits required for the high cadence survey must include the ditherFactor
    nHighVisitsPerYear = nHighPointings * ditherFactor * nVisitsPerFieldPerYear
    if verbose: print('Number of visits per year to high cadence region: ',nHighVisitsPerYear)
    if nHighVisitsPerYear > nVisitsPerYear:
        nHighVisitsPerYear = nVisitsPerYear
        return 0, 0.0
    
    # Number of visits remaining for the low cadence region, again factoring in dithering
    nLowVisitsPerYear = max((nVisitsPerYear - nHighVisitsPerYear),0)
    if nLowVisitsPerYear > 0:
        nLowVisitsPerYear /= 6
    if verbose: print('Number of visits remaining for low cadence region: ',nLowVisitsPerYear)
    
    # Calculate the cadence that can be achieved with this many visits, assuming uniform sampling:
    lowCadenceArea = len(lowCadencePixels)*pixArea
    nLowPointings = round(lowCadenceArea / AreaPerVisit, 0)
    if verbose: print('Number of pointings in low cadence region: ',nLowPointings)
    nVisitsPerLowPointing = nLowVisitsPerYear / nLowPointings
    if verbose: print('Number of visits per pointing in low cadence region: ',nVisitsPerLowPointing)
    lowCadence = nightsObsSeason / nVisitsPerLowPointing
    if verbose: print('Low cadence achieved: ',lowCadence)
    
    return nLowVisitsPerYear, lowCadence
```

```python
highCadence = np.arange(2.0,10.0,0.25)
lowVisits = np.zeros(len(highCadence))
lowCadence = np.zeros(len(highCadence))

for i,cadence in enumerate(highCadence):
    lv,lc = estimate_survey_low_cadence(highCadencePixels,lowCadencePixels,cadence,NTotalVisits)
    lowVisits[i] = lv
    lowCadence[i] = lc

(fig, (ax1)) = plt.subplots(1, 1, figsize=(10, 10))
ax1.plot(highCadence, lowCadence, 'r-')
plt.xlabel('High cadence region interval [days]', fontsize=16)
plt.ylabel('Low cadence region interval [days]', fontsize=16)
plt.title('Interval between sequential observations (all filters combined) for low/high cadence regions')
plt.xticks(fontsize=16)
plt.yticks(fontsize=16)
```

A cadence of about 5-10 days would be acceptable for the low cadence region in order to maintain a cadence around 2-3days for the high cadence region, and would enable most science cases.  

It's also possible to optimize this over the size of the low-cadence region, as well as the cadence achieved for both regions:

```python
highCadence = np.arange(2.0,10.0,0.25)
GPthreshold = np.arange(0.01,1.5,0.1)
highVisits = np.zeros((len(highCadence),len(GPthreshold)))
lowSurveyArea = np.zeros((len(highCadence),len(GPthreshold)))
lowCadence = np.zeros((len(highCadence),len(GPthreshold)))

pixArea = hp.nside2pixarea(NSIDE, degrees = True)

for i,cadence in enumerate(highCadence):
    for j, threshold in enumerate(GPthreshold):
        lowCadenceRegion = defineLowCadenceRegion(galSciencePriorityMapTable, visiblePixels, 
                                                    highCadencePixels, priorityThresholds=[threshold, 0.00005])
        lowCadencePixels = np.where(lowCadenceRegion > 0.0)[0]
        lv,lc = estimate_survey_low_cadence(highCadencePixels,lowCadencePixels,cadence,NTotalVisits,verbose=False)
        highVisits[i,j] = cadence
        lowSurveyArea[i,j] = len(lowCadencePixels)*pixArea
        lowCadence[i,j] = lc
                                                  
fig = plt.figure()
ax = plt.axes(projection='3d')
ax.contour3D(highVisits, lowSurveyArea, lowCadence, 150, cmap='binary')
ax.set_xlabel('High cadence region interval [days]')
ax.set_ylabel('Low cadence survey area [deg$^{2}$]')
ax.set_zlabel('Low cadence region interval [days]')
ax.view_init(20, 30)
```
