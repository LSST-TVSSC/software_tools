# Spatial Distribution of Microlensing Events

Rubin's combination of faint limiting magnitude, long-baseline and wide-area survey capability are some the key factors that distinguish it from previous microlensing surveys, and present unique science drivers.  
1) The detection of microlensing events over all timescales, from ~10days to > 1yr, including low mass stellar lenses as well as compact objects and black holes
2) The exploration of these populations in different regions of the galaxy outside the Galactic Bulge.

(Note: planetary events and the synergies with the Roman survey of the Galactic Bulge are also of great interest but confined to a limited spatial area, so these are considered separately). 

These science drivers require us to sample a range of lines of sight through the galactic population, which in terms of survey strategy translates to timeseries imaging of fields distributed across a range of galactic longitude, in fields where the stellar density is highest, particularly through the arms of the Milky Way and towards the Magellenic Clouds.  

The goal of this notebook is to explore how well different OpSims serve these goals, drawing on the [https://github.com/lsst/rubin_sim/blob/main/rubin_sim/maf/mafContrib/microlensingMetric.py] Microlensing Metric developed by N. Abrams, M.P.G. Hundertmark and the TVS Microlensing group, in combination with the [https://github.com/LSST-TVSSC/software_tools/tree/main/footprint_maps] priority maps for galactic science by R. Street et al. 

```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import rubin_sim.maf as maf
from rubin_sim.data import get_data_dir
import healpy as hp
from astropy import units as u
from astropy_healpix import HEALPix
from astropy.coordinates import Galactic, TETE, SkyCoord
from astropy.io import fits

NSIDE = 64
NPIX = hp.nside2npix(NSIDE)
```

We'll start by looking at the Baseline v2.0 OpSim as a reference:

```python
from rubin_sim.data import get_baseline

opsim_fname = get_baseline()
print(opsim_fname)

runName = os.path.split(opsim_fname)[-1].replace('.db', '')
print(runName)
opsim_db = maf.OpsimDatabase(opsim_fname)
```

By design, the Microlensing Metric produces a summary metric calculated over all HEALpixels, but here we want to examine the spatial distribution information.  To do this we experiment with a customized HEALpixel dataSlicer.

```python
def HEALpixMicrolensingSlicer(t0,u0,tE,
    n_events=10000,
    nside=128,
    filtername="r",
):
    """
    Generate a UserPointSlicer with a population of microlensing events, suitable
    for use with the MicrolensingMetric.

    This dataSlicer has been adapted from code by Peter Yoachim, with customizations
    to generate a catalog of events for a specific region of the sky rather than
    a uniform random distribution of pointings.

    Parameters
    ----------
    desired_healpix : array of int
        HEALpixel indices of the desired survey region
    min_crossing_time : float (1)
        The minimum crossing time for the events generated (days)
    max_crossing_time : float (10)
        The max crossing time for the events generated (days)
    t_start : float (1)
        The night to start generating peaks (days)
    t_end : float (3652)
        The night to end generating peaks (days)
    n_events : int (10000)
        Number of microlensing events to generate
    nside : int (128)
        HEALpix nside, used to pick which stellar density map to load
    filtername : str ('r')
        The filter to use for the stellar density map
    """

    # We simulate a single event for each HEALpixel in the desired survey region,
    # owing to the constraints on how HEALpixel slicers are computed.
    # A set of events is accumulated by repeating the calculation for over the map
    slicer = maf.slicers.HealpixSlicer(nside=NSIDE, useCache=False)

    # Add the additional microlensing information about each object to the slicer.
    # In effect this simulates the same event happening at every HEALpix on the sky
    slicer.slicePoints["peak_time"] = np.array( [t0] )
    slicer.slicePoints["crossing_time"] = np.array( [tE] )
    slicer.slicePoints["impact_parameter"] = np.array( [u0] )

    return slicer
```

The following code is designed to calculate the Microlensing Metric with this customized dataSlicer, with configurable event parameters. 

```python
def runMicrolensingMetric(tE,t0,u0,opsim_db,runName):
    bundleList = []

    constraint = 'fiveSigmaDepth > 21.5'
    plotDict = {'colorMax': 950}
    
    metric1 = maf.mafContrib.microlensingMetric.MicrolensingMetric(metricCalc="detect")
    #metric2 = maf.mafContrib.microlensingMetric.MicrolensingMetric(metricCalc="Npts")
    
    slicer = HEALpixMicrolensingSlicer(t0,u0,tE,n_events=1,nside=NSIDE,filtername="r")
    
    bundleList.append( maf.MetricBundle(metric1, slicer, None, runName=runName,
                                                plotDict=plotDict, 
                                                info_label=f'tE {tE} days') )
   # bundleList.append( maf.MetricBundle(metric2, slicer, None, runName=runName,
    #                                            plotDict=plotDict, 
     #                                           info_label=f'tE {tE} days') )
    bundleDict = maf.metricBundles.makeBundlesDictFromList(bundleList)
    bundleGroup = maf.MetricBundleGroup(bundleDict, opsim_db, outDir='test', resultsDb=None)
    bundleGroup.runAll()
    
    return bundleDict
```

```python
def extract_metric_data(bundleDict, runName, tE):
    outputName = runName.replace('.','_')+'_MicrolensingMetric_detect_tE_'+str(round(tE,1)).replace('.','_')+'_days_HEAL'
    metricData = bundleDict[outputName].metricValues.filled(0.0)
    return metricData
```

We start by looking at events with Einstein Crossing Times, tE, typical of main sequence stars in the Milky Way disk.
We simulate a sample of events with a realistic range in parameters for each HEALpixel, with the same event parameters used in each case.  The Microlensing Metric then simulates events with these parameters per HEALpixel and determines how many of them would have been detected, given the survey observation timestamps per HEALpixel from the OpSim.  By summing these results over the sample of events, we can test how well events of this timescale would be recovered as a function of position. 

#### Important caveats: 
1) the MicrolensingMetric by default considers data in a single-filter lightcurve, which significantly impacts the sampling in LSST, where they sample through all filters by dafault.  Since microlensing observations can effectively combine lightcurves in different filters, this creates an under-estimate. 

2) In a full simulation, you would want to simulate more events per pixel for better statistical sampling

3) I'm using a single, typical stellar-mass lens event here as an example; for a full exploration you would need to look at a greater range of events

```python
nevents_per_hp = 10
tE = 30.0

t_start=1
t_end=3652

metricMap = np.zeros(NPIX)
for i in range(0,nevents_per_hp,1):
    t0 = np.random.uniform(low=t_start, high=t_end, size=1)[0]
    u0 = np.random.uniform(low=0.001, high=0.5, size=1)[0]
    print('Simulated event parameters: ',tE,t0,u0)
    
    bundleDict = runMicrolensingMetric(tE,t0,u0,opsim_db,runName)
    
    metricData = extract_metric_data(bundleDict, runName, tE)
    metricMap += metricData
```

```python
hp.mollview(metricMap/nevents_per_hp, title='Fraction of events detected')
hp.graticule()
```

We can now evaluate how well we recover events as a function of spatial position.

For example, we can calculate the fraction of events recovered for each HEALpixel within the pencilbeam fields proposed for the Galactic Plane survey, since these are deliberately distributed across galactic longitude.

The HEALpixel locations are recorded in the footprint map data table.

```python
science_map_file = '/Users/rstreet1/software/LSST-TVS_software_tools/footprint_maps/priority_GalPlane_footprint_map_data_sum.fits'
with fits.open(science_map_file) as hdul:
    mapTable = hdul[1].data

pencilbeams = mapTable['pencilbeams_map']
```

```python
hp.mollview(pencilbeams, title='Pencilbeam fields')
hp.graticule()
```

```python
idx = np.where(pencilbeams > 0.0)[0]
detfrac_pencilbeams = np.zeros(NPIX)
detfrac_pencilbeams[idx] = metricMap[idx]/nevents_per_hp
```

We now calculate the galactic longitude corresponding to the centre of each HEALpixel, in order to plot the detection fraction as a function of l

```python
def calc_healpixel_coordinates(healpixels):
    """
    Calculate the l,b of each HEALpixel center from the list of HEALpixels
    """

    pixelPositions = []
    (theta, phi) = hp.pix2ang(NSIDE,range(0,len(healpixels),1))

    ra = np.rad2deg(phi)
    dec = np.rad2deg((np.pi/2.0) - theta)

    skyPositions = SkyCoord(ra*u.deg,
                             dec*u.deg,
                             frame='icrs')
    skyGalactic = skyPositions.transform_to(Galactic)

    return skyGalactic
```

```python
hpcoord = calc_healpixel_coordinates(detfrac_pencilbeams)
gal_long = np.array([x.l.deg for x in hpcoord])
order = np.argsort(gal_long)

fig = plt.figure(1,(10,10))
plt.plot(gal_long[order],detfrac_pencilbeams[order],'b-')
plt.xlabel('Galactic longitude [deg]')
plt.ylabel('Event detection fraction')
```

A simple metric to evaluate how well microlensing events are detected at a range of longitudes is to evaluate the mean and standard deviation of the detection rate across all of these pencilbeam HEALpixels

```python
beam_pixels = np.where(detfrac_pencilbeams > 0.0)
print('Mean = '+str(detfrac_pencilbeams[beam_pixels].mean())+', StdDev = '+str(detfrac_pencilbeams[beam_pixels].std()))
```
