# OpSim Database Structure and Content

Typically we interact with opsim databases using the MAF framework which compute the metrics.  However, for some use cases we want to access the row information in the database itself, so this notebook explores the available content. 

```python
from os import path
import rubin_sim.maf as maf
import numpy as np 
import sqlite3 
import pandas as pd
import healpy as hp
import matplotlib.pyplot as plt

NSIDE = 64
```

```python
# UPDATE THIS TO THE LATEST DB
DATA_DIR = '/Users/rstreet/rubin_sim_data/sim_baseline' 
DB_FILE = 'baseline_v4.3.1_10yrs.db'

ops_db_path = path.join(DATA_DIR, DB_FILE)

run_name = path.split(path.basename(ops_db_path))[-1].replace('.db', '')
```

```python
db = sqlite3.connect(ops_db_path)
obs_table = pd.read_sql_query("SELECT * from observations", db)
for col in obs_table.columns:
    print(col)
```
